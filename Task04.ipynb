```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/US_Accidents_March23.csv', nrows=50000)

df.head()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour
```

```python
import matplotlib.pyplot as plt

df['Hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Accidents by Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
df['Weather_Condition'].value_counts().head(10).plot(kind='bar')
plt.title('Top Weather Conditions')
plt.show()
```

```python
df['Visibility(mi)'].plot(kind='hist', bins=30)
plt.title('Visibility vs Accidents')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x=df['Start_Lng'], y=df['Start_Lat'], s=1)
plt.title('Accident Hotspots')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Severity'].value_counts().sort_index().plot(kind='bar')
plt.title('Accident Severity Distribution')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()
```

📌 Insights:

*Most accidents happen during morning and evening peak hours

*Weather conditions like rain, fog, and cloudy weather increase accidents

*Low visibility leads to higher accident probability

*Certain areas show dense accident clusters (hotspots)
